# Unemployment Analysis with Python

This project analyzes unemployment data using Python and visualization techniques.

## Objective

The objective of this project is to analyze unemployment trends and visualize unemployment rates using Python libraries.

### Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the Dataset

```python
df1 = pd.read_csv('/content/Unemployment in India.csv')

df2 = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
```

```python
df1.head()
```

```python
df2.head()
```

### Checking Dataset Information

```python
print(df1.info())

print(df1.describe())

print(df1.isnull().sum())
```

```python
print(df2.info())

print(df2.describe())

print(df2.isnull().sum())
```

### Dataset Columns

```python
print(df1.columns)

print(df2.columns)
```

### Unemployment Rate Distribution

```python
plt.figure(figsize=(10,6))

sns.histplot(df2.iloc[:,3], kde=True)

plt.title("Unemployment Rate Distribution")

plt.show()
```

### Unemployment Rate by Region

```python
plt.figure(figsize=(12,6))

sns.barplot(x=df2.iloc[:,0], y=df2.iloc[:,3])

plt.xticks(rotation=90)

plt.title("Unemployment Rate by Region")

plt.show()
```

### Correlation Heatmap

```python
numeric_df = df2.select_dtypes(include=['float64', 'int64'])

sns.heatmap(numeric_df.corr(), annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

### Average Unemployment Rate

```python
average_rate = df2.iloc[:,3].mean()

print("Average Unemployment Rate:", average_rate)
```

## Conclusion

The unemployment dataset was successfully analyzed using Python. Different visualizations were created to understand unemployment trends and unemployment rates across regions.
